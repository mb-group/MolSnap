```python
import duckdb
```

```python
def get_compounds_by_id(id_type, id_value):
    """
    id_type: 'dois', 'pcmid', or 'pmids'
    """
    con = duckdb.connect(database=':memory:')
    
    query = f"""
    SELECT m.CID
    FROM '../data/papers.parquet' l
    JOIN '../data/mapping.parquet' m ON l.PCLID = m.PCLID
    WHERE l.{id_type} = ?
    """
    
    return con.execute(query, [id_value]).df()

# Example Usage
print(get_compounds_by_id('dois', '10.1007/bf01175560'))
```

```python
print(get_compounds_by_id('pmids', '469572'))
```
