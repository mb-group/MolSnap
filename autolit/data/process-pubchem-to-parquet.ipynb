```python
import pandas as pd
```

```python
def df_to_parquet(reader, output_file: str): 
    for i, chunk in enumerate(reader):
        # Write first chunk to create file, append others
        mode = 'w' if i == 0 else 'a'
        chunk.to_parquet(output_file, engine='fastparquet', append=(i > 0))
```

```python
df_cid_lit = pd.read_csv("raw/cid_literature.tsv", sep="\t", chunksize=100_000, low_memory=False)
```

```python
df_to_parquet(df_cid_lit, "mapping.parquet")
```

```python
df_lit_cid = pd.read_csv("raw/literature_linked_to_cid.tsv", sep="\t", chunksize=100_000, low_memory=False)
```

```python
df_to_parquet(df_lit_cid, "papers.parquet")
```
